```python
df = pd.read_csv("Task-1(skillcraft).csv")
year_columns = [col for col in df.columns if col.isdigit()]
```

```python
df = df.dropna(subset=['Country Name', 'Indicator Name'])
df = df[df['Country Name'].apply(lambda x: isinstance(x, str))]
df = df[df['Indicator Name'].apply(lambda x: isinstance(x, str))]
```

```python
country_options = sorted(df['Country Name'].unique())
indicator_options = sorted(df['Indicator Name'].unique())
```

```python
def plot_bar(country, indicator):
    data = df[(df['Country Name'] == country) & (df['Indicator Name'] == indicator)]

    if data.empty:
        print("No data available for this combination.")
        return

    melted = data.melt(id_vars=['Country Name', 'Indicator Name'],value_vars=year_columns,var_name='Year', value_name='Value')

    melted['Year'] = pd.to_numeric(melted['Year'], errors='coerce')
    melted.dropna(subset=['Value'], inplace=True)

    plt.figure(figsize=(12,6))
    plt.bar(melted['Year'], melted['Value'], color='#4682B4')
    plt.title(f"{indicator} in {country}", fontsize=14)
    plt.xlabel("Years")
    plt.ylabel("Population in Billions")
    plt.xticks(melted['Year'], rotation=90, fontsize=8)
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.tight_layout()
    plt.show()


interact(plot_bar, country=country_options, indicator=indicator_options)
```
